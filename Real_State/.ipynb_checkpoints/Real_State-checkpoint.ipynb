```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Real estate.csv")
```

```python
df.head()
```

```python
df=df.drop("No",axis=1)
```

```python
df.columns=["transaction date","house age","distance from MRT","number of conv stores","lat","long","prices per unit"]
```

```python
df.info()
```

```python
df["transaction date"]=[int(str(x)[:4]) for x in df["transaction date"]]
```

```python
df.describe()
```

```python
sns.histplot(df["transaction date"])
```

```python
sns.histplot(df["house age"],kde=True)
```

```python
sns.histplot(df["distance from MRT"],kde=True)
```

```python
sns.histplot(x=df["number of conv stores"])
```

```python
sns.histplot(df["long"])
```

```python
sns.histplot(df["lat"])
```

```python
sns.histplot(df["prices per unit"],kde=True)
```

```python
sns.boxplot(x=df["transaction date"],y=df["prices per unit"])
```

```python
sns.jointplot(y=df["house age"],x=df["prices per unit"])
```

```python
sns.jointplot(x=df["distance from MRT"],y=df["prices per unit"],kind="hist")
```

# What is the preportion of genders

# How is the distribution of age

```python
sns.histplot(data=data["Age"],bins=50,color="green")
```

# How is the distribution of scores

```python
sns.histplot(data=data["Spending Score (1-100)"],bins=50,color="gray")
```

# What is the score based on age

```python
plt.figure(figsize=(15,6))
sns.barplot(data=data,x="Age",y="Spending Score (1-100)")
plt.show()
```

**So the mean highest given score is for 39s and the absolut highest score is  for 33s**

**the mean least score is for 64s and the the least absolute score is given bu 37s**

**The most variation in score is for 33 means most of them were not unanimous**

**The least variation in score is for 70 means most of them were unanimous**

**I did not mention 64 because there is only one person at the age of 64 and the low score of him or her is unreasonable, observing other scores.
 But it is the absolut lowest score**

# Which company is more expensive

```python
plt.figure(figsize=(25,5))
sns.barplot(data=data,x="competitorname",y="pricepercent")
plt.xticks(rotation=90)
plt.show()
```

# Which company puts more wine

```python
plt.figure(figsize=(25,5))
sns.barplot(data=data,x="competitorname",y="winpercent")
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("/content/south-korean-pollution-data.csv")
```

```python
data.head()
```

```python
data=data.drop("Unnamed: 0",axis=1)
```

```python
data.info()
```

```python
year=[]
month=[]
day=[]

for i in data["date"]:
  yeartemp, monthtemp, daytemp= i.split("/")
  year.append(int(yeartemp))
  month.append(int(monthtemp))
  day.append(int(daytemp))
```

```python
data["year"]=year
data["month"]=month
data["day"]=day
```

```python
data=data.drop("date", axis=1)
```

```python
data.describe()
```

```python
def Numerizer(column):
  temp=[]
  for i in column:
    if i != "" and i != " ":
      temp.append(int(i))
    else:
      temp.append(-1)
  return temp
```

```python
data[" pm25"]=Numerizer(data[" pm25"])
data[" pm10"]=Numerizer(data[" pm10"])
data[" o3"]=Numerizer(data[" o3"])
data[" no2"]=Numerizer(data[" no2"])
data[" so2"]=Numerizer(data[" so2"])
data[" co"]=Numerizer(data[" co"])
```

```python
for i in [' pm25', ' pm10', ' o3', ' no2', ' so2', ' co']:
  for j in data.index:
    if data[i][j]==-1:
      data=data.drop(j, axis=0) 
```
